# Extracting spectrograms

```python
frame_size = 2048
hop = 512
```

# Extracting Mel spectrograms

```python
# compute the Short time fourier transform of the signal
stft = librosa.stft(sample, n_fft = frame_size, hop_length = hop)
```

```python
stft.shape
# THE FIRST DIMENSION IS EQUAL TO: (frame_size / 2) + 1
# SECOND DIMENSION IS: The number of frames, or temporal bins
```

```python
# The output of an stft are a series of complex coefficients
print(type(stft[0][0]))
print(stft[0][0])
```

```python
# converting from the stft to a spectrogram

# calculate the 'squared magnitude'
scaled = np.abs(stft) ** 2 
scaled.shape
```

```python
print(type(scaled[0][0]))
```

```python
def spectrogram(data, sr = 44100, hop = 512, y_axis = "linear"):
    plt.figure(figsize = (25, 10))
    librosa.display.specshow(data, sr = sr, hop_length = hop, x_axis = 'time', y_axis = y_axis)
```
